```python
class Nasabah:
    def __init__(self,nama,alamat,no_telepon):
        self.nama = nama
        self.alamat = alamat
        self.no_telepon = no_telepon
    def showInfo(self):
        print("\n\tNama: {} \n\tAlamat: {} \n\tno_telepon: {} ".format(self.nama,self.alamat,self.no_telepon))
class Rekening(Nasabah):
    def __init__(self,nama,alamat,no_telepon,no_rekening,saldo):
        self.no_rekening = no_rekening
        self.saldo = saldo
        super().__init__(nama,alamat,no_telepon)
        print("\n\tno_rekening: {} \n\tsaldo: {}".format(self.no_rekening,self.saldo))
    def tambah_saldo(self):
        self.uang = int(input("Masukkan jumlah saldo yang akan ditambahkan[200000]: "))
        self.saldo = int(self.saldo)+int(self.uang)
        print("Saldo Anda Saat ini Adalah: ",self.saldo)
    def ambil_saldo(self):
        self.uang = int(input("Masukkan Jumlah saldo yang akan diambil[200000]: "))
        self.saldo = int(self.saldo)-int(self.uang)
        print("Saldo Anda Saat ini Adalah: ",self.saldo)
    def pilihan(self):
        while True :
            pilihan = input("\n\tMasukkan Pilihan:")
            if pilihan == "tambah saldo":
                self.tambah_saldo()
            if pilihan == "ambil saldo":
                self.ambil_saldo()
            else:
                print("pilihan yang anda masukkan salah")
            a = input("lagi[y/t]")
            b = "t"
            if a == b:
                break
n1 = Nasabah("Viki","Sampang","085807160919")
n1.showInfo()
r1= Rekening("Viki","Sampang","085807160919",12345678,500000)
r1.pilihan()
```

```python
class Lingkaran:
    phi = 3.14
    def __init__(self,jari_jari):
        self.jari_jari = jari_jari
    def tampilkan(self):
        print("jari-jari: ",self.jari_jari)
    def luas_permukaan(self):
        print("luas permukaan: ",4/self.phi*self.jari_jari**2)
    def luas_lingkaran(self):
        print("luas lingkaran: ",self.phi*self.jari_jari**2)
    def showInfo(self):
        print("keliling lingkaran: ",2*self.phi*self.jari_jari**2)  
class Tabung(Lingkaran):
    def __init__(self,jari_jari,tinggi_tabung):
        self.tinggi_tabung = tinggi_tabung
        super().__init__(jari_jari)
        print("tinggi tabung: {}".format(self.tinggi_tabung))
    def showInfo(self):
        print("Luas Permukaan Tabung: ",2*self.phi*self.jari_jari*(self.tinggi_tabung+self.jari_jari))
        
L1 = Lingkaran(10)
L1.tampilkan()
L1.luas_permukaan()
L1.luas_lingkaran()
L1.showInfo()
T1 = Tabung(10,12)
T1.showInfo()
```

```python
def fpb(a,b):
    if b == 0:
        return a
    else:
        return fpb(b,a%b)
class Pecahan:
    def __init__(self,pembilang,penyebut):
        self.pembilang = pembilang
        self.penyebut = penyebut
    def showInfo(self):
        print("\n\tpembilang: {} \n\tpenyebut {}".format(self.pembilang,self.penyebut))
    def sederhana(self):
        print("\n\tpecahan: "+str(int(self.pembilang))+"/"+str(int(self.penyebut)))
        self.show_pembilang = self.pembilang/fpb(self.pembilang,self.penyebut)
        self.show_penyebut = self.penyebut/fpb(self.pembilang,self.penyebut)
        print("\n\tpecahan sederhana: "+str(int(self.show_pembilang))+"/"+str(int(self.show_penyebut)))
P1 = Pecahan (9,6) 
P1.showInfo()
P1.sederhana()
P2 = Pecahan (24,32) 
P2.showInfo()
P2.sederhana()

```

```python
class Katagori:
    def __init__(self,nama_katagori):
        self.nama_katagori = nama_katagori
class Barang(Katagori):
    def __init__(self,nama_barang,kode,harga_satuan):
        self.nama_barang = nama_barang
        self.kode = kode
        self.harga_satuan = harga_satuan
class Pembeli:
    def __init__(self,nama_pembeli):
        self.nama_pembeli = nama_pembeli
class TransaksiPembelian(Barang,Pembeli):
    def __init__(self,nama_katagori,nama_barang,kode,harga_satuan,nama_pembeli):
        self.nama_pembeli = nama_pembeli
        Katagori.__init__(self,nama_katagori)
        Barang.__init__(self,nama_barang,kode,harga_satuan)
        print("\n\tkatagori: {}  \n\tnama barang: {} \n\tkode: {} \n\tharga satuan: {} \n\tnama pembeli: {}".format(
            self.nama_katagori,
            self.nama_barang,
            self.kode,
            self.harga_satuan,
            self.nama_pembeli))
P1 = Pembeli("Viki")
T1 = TransaksiPembelian("Gatsby","Pomade","122","27000","Viki")
```

```python
class Barang:
    def __init__(self,nama,harga,diskon):
        self.nama = nama
        self.harga = harga
        self.diskon = diskon
    def showInfo(self):
        #Abstract Class
        raise NotImplementedError("Subclass must implement abstract method")
class Makanan(Barang):
    def __init__(self,nama,harga,diskon,exdate):
        self.exdate = exdate
        super().__init__(nama,harga,diskon)
    def showInfo(self):
        print("nama: {} harga: {} diskon: {} exdate: {}".format(
            self.nama,
            self.harga,
            self.diskon,
            self.exdate))
class Pakaian(Barang):
    def __init__(self,nama,harga,diskon,ukuran):
        self.ukuran = ukuran
        super().__init__(nama,harga,diskon)
    def showInfo(self):
        print("nama: {} harga: {} diskon: {} ukuran: {}".format(
            self.nama,
            self.harga,
            self.diskon,
            self.ukuran))
Obj1 = Makanan("Geprek",8000,"10%",2)
Obj2 = Pakaian("volcom",200000,"15%","XL")
for Obj in (Obj1,Obj2):
    Obj.showInfo()
```

```python
class KatagoriBarang:
    def __init__(self,kode,deskripsi):
        self.kode = kode
        self.deskripsi = deskripsi
    def showInfo(self):
        print("kode: {} deskripsi: {}".format(self.kode,self.deskripsi))
class Barang(KatagoriBarang):
    def __init__(self,kode,deskripsi,nama,harga,diskon):
        self.nama = nama
        self.harga = harga
        self.diskon = diskon
        super().__init__(kode,deskripsi)
    def showInfo(self):
        print("nama: {} kode: {} deskripsi: {} harga: {} diskon: {}".format(
            self.nama,
            self.kode,
            self.deskripsi,
            self.harga,
            self.diskon))
class Makanan(Barang):
    def __init__(self,nama,kode,deskripsi,harga,diskon,exdate):
        self.exdate = exdate
        super().__init__(kode,deskripsi,nama,harga,diskon)
    def showInfo(self):
        print("nama: {} kode: {} deskripsi: {} harga: {} diskon: {} exdate: {}".format(
            self.nama,
            self.kode,
            self.deskripsi,
            self.harga,
            self.diskon,
            self.exdate))
class Pakaian(Barang):
    def __init__(self,nama,kode,deskripsi,harga,diskon,ukuran):
        self.ukuran = ukuran
        super().__init__(kode,deskripsi,nama,harga,diskon)
    def showInfo(self):
        print("nama: {} kode: {} deskripsi: {} harga: {} diskon: {} exdate: {}".format(
            self.nama,
            self.kode,
            self.deskripsi,
            self.harga,
            self.diskon,
            self.ukuran))
#Implementasi menggunakan polymorpism with function
def Beli(barang):
    barang.showInfo()
Obj1 = Makanan("Geprek",122,"Krispi",8000,"10%",2)
Obj2 = Pakaian("volcom",123,"Kain Catton",200000,"15%","XL")
Beli(Obj1)
Beli(Obj2)
#Implementasi menggunakan polymorpism with class
for Barang in (Obj1,Obj2):
    Barang.showInfo()
```
